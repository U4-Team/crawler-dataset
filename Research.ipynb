# Dataset stats

```python
%matplotlib inline
import pandas
from matplotlib.pyplot import hist
from collections import Counter
from matplotlib.pyplot import figure

dataframe = pandas.read_excel('./dataset.xlsx')

okved_count = Counter(dataframe['ОКВЭД_Наименование'])
okved_count_df = pandas.DataFrame.from_dict(okved_count, orient='index')

inn_count = Counter(dataframe['ИНН'])
inn_count_df = pandas.DataFrame.from_dict(inn_count, orient='index')

msp_type_count = Counter(dataframe['тип МСП: 1-ЮЛ, 2-ИП'])
msp_type_count_df = pandas.DataFrame.from_dict(msp_type_count, orient='index')

msp_cat_count = Counter(dataframe['категория МСП'])
msp_cat_count_df = pandas.DataFrame.from_dict(msp_cat_count, orient='index')

pp_count = Counter(dataframe['ПП'])
pp_count_df = pandas.DataFrame.from_dict(pp_count, orient='index')

spark_id_count = Counter(dataframe['SparkID'])
spark_id_count_df = pandas.DataFrame.from_dict(spark_id_count, orient='index')

reg_date_count = Counter(dataframe['Дата регистрации'])
reg_date_count_df = pandas.DataFrame.from_dict(reg_date_count, orient='index')

phone_count = Counter(dataframe['Телефоны'])
phone_count_df = pandas.DataFrame.from_dict(phone_count, orient='index')

email_count = Counter(dataframe['EMAIL'])
email_count_df = pandas.DataFrame.from_dict(email_count, orient='index')

site_count = Counter(dataframe['Сайт'])
site_count_df = pandas.DataFrame.from_dict(site_count, orient='index')

address_count = Counter(dataframe['Юридический адрес'])
address_count_df = pandas.DataFrame.from_dict(address_count, orient='index')
```
